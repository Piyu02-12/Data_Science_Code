# Model Building

```python
model=smf.ols("Profit~RDS+ADM+MKTS" , data=data1).fit()
```

# Model Testing

```python
#finding coefficient parameter
model.params
```

```python
#finding t-value and p-value
model.tvalues , np.round(model.pvalues , 5)
```

```python
#finding rsquared values
model.rsquared , model.rsquared_adj  #model accuracy is 94.75%
```

```python
#building slr and mlr model for insignificant values ADM & MKTS 
```

```python
slr_a =smf.ols("Profit~ADM" , data=data1).fit()
slr_a.tvalues ,  slr_a.pvalues                      #ADM has insignificant pvalue
```

```python
slr_m = smf.ols("Profit~MKTS" , data = data1).fit()
slr_m.tvalues , slr_m.pvalues                        #MKTS has significant pvalues
```

```python
mlr_am = smf.ols("Profit~ADM+MKTS",data=data1).fit()
mlr_am.tvalues , mlr_am.pvalues                      #has significant pvalues
```

```python
#residual analysis
sm.qqplot(model.resid,line='q')
plt.title("normal qqplot of residual")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
#test for homoscedasticity or heteroscedasticity
def standard_values(vals) : return(vals-vals.mean())/vals.std()
```

```python
plt.scatter(standard_values(model.fittedvalues ), standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show() 
```
